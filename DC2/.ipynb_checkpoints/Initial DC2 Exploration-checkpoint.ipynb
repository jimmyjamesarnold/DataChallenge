I can label these users using a SQL case statement. 
Specifically, I want label user_id's that were active in July, BUT NOT active in August. Those will be my 'stopped' group.
For comparison, I'll label the user_id's that were active in July, AND active in August. Those will be my 'engaged' group.

Then, I can do some comparisons to see what's distinct about the stopped group vs the engaged users.

```python
# first, let's get the number of events, per user, per month.

sql_query = """
SELECT user_id, 
        COUNT(event_name) AS monthly_events,
        EXTRACT('month'  FROM occurred_at) AS month
 FROM events
 WHERE event_type = 'engagement'
 GROUP BY 1, 3
"""
data_from_sql = pd.read_sql_query(sql_query,con)
data_from_sql.head()
```

```python
# next, I'm going to pivot the monthly events to columns doing a sub-query

sql_query = """
SELECT user_id,
        SUM(CASE WHEN month = 5.0 THEN monthly_events ELSE NULL END) AS may,
        SUM(CASE WHEN month = 6.0 THEN monthly_events ELSE NULL END) AS jun,
        SUM(CASE WHEN month = 7.0 THEN monthly_events ELSE NULL END) AS jul,
        SUM(CASE WHEN month = 8.0 THEN monthly_events ELSE NULL END) AS aug
 FROM(
        SELECT user_id, 
                COUNT(event_name) AS monthly_events,
                EXTRACT('month'  FROM occurred_at) AS month
         FROM events
         WHERE event_type = 'engagement'
         GROUP BY 1, 3
         ) sub
 GROUP BY 1
"""
data_from_sql = pd.read_sql_query(sql_query,con)
data_from_sql.head()
```

```python
# OK, I can use this table to create the labels for the 'stopped' and 'engaged' groups

sql_query = """
SELECT *
FROM (
    SELECT user_id,
            SUM(CASE WHEN month = 5.0 THEN monthly_events ELSE NULL END) AS may,
            SUM(CASE WHEN month = 6.0 THEN monthly_events ELSE NULL END) AS jun,
            SUM(CASE WHEN month = 7.0 THEN monthly_events ELSE NULL END) AS jul,
            SUM(CASE WHEN month = 8.0 THEN monthly_events ELSE NULL END) AS aug
     FROM(
            SELECT user_id, 
                    COUNT(event_name) AS monthly_events,
                    EXTRACT('month'  FROM occurred_at) AS month
             FROM events
             WHERE event_type = 'engagement'
             GROUP BY 1, 3
             ) sub
     GROUP BY 1
     ) monthly_engagement
 WHERE jul > 0 AND aug IS NULL
"""
stopped = pd.read_sql_query(sql_query,con)
stopped.head()
```

```python
stopped.shape
```

```python
# OK, I can use this table to create the 'stopped' and 'engaged' groups

sql_query = """
SELECT user_id, 
    CASE WHEN jul > 0 AND aug IS NOT NULL THEN 'engaged' 
         WHEN jul > 0 AND aug IS NULL THEN 'stopped'
         ELSE NULL END AS target_group
FROM (
    SELECT user_id,
            SUM(CASE WHEN month = 5.0 THEN monthly_events ELSE NULL END) AS may,
            SUM(CASE WHEN month = 6.0 THEN monthly_events ELSE NULL END) AS jun,
            SUM(CASE WHEN month = 7.0 THEN monthly_events ELSE NULL END) AS jul,
            SUM(CASE WHEN month = 8.0 THEN monthly_events ELSE NULL END) AS aug
     FROM(
            SELECT user_id, 
                    COUNT(event_name) AS monthly_events,
                    EXTRACT('month'  FROM occurred_at) AS month
             FROM events
             WHERE event_type = 'engagement'
             GROUP BY 1, 3
             ) sub
     GROUP BY 1
     ) monthly_engagement
"""
user_labels = pd.read_sql_query(sql_query,con)
user_labels.head()
```

```python
# first, let's keep it contained to the events table.
# I'm going to self-join with the new labels I made, generate a df,
# and do some EDA.

sql_query = """
SELECT labels.*, events.*
FROM (
    SELECT user_id, 
        CASE WHEN jul > 0 AND aug IS NOT NULL THEN 'engaged' 
             WHEN jul > 0 AND aug IS NULL THEN 'stopped'
             ELSE NULL END AS target_group
    FROM (
        SELECT user_id,
                SUM(CASE WHEN month = 5.0 THEN monthly_events ELSE NULL END) AS may,
                SUM(CASE WHEN month = 6.0 THEN monthly_events ELSE NULL END) AS jun,
                SUM(CASE WHEN month = 7.0 THEN monthly_events ELSE NULL END) AS jul,
                SUM(CASE WHEN month = 8.0 THEN monthly_events ELSE NULL END) AS aug
         FROM(
            SELECT user_id, 
                    COUNT(event_name) AS monthly_events,
                    EXTRACT('month'  FROM occurred_at) AS month
             FROM events
             WHERE event_type = 'engagement'
             GROUP BY 1, 3
             ) sub
         GROUP BY 1
         ) sub_sub
) labels
LEFT JOIN events
 ON labels.user_id = events.user_id
 WHERE target_group IS NOT NULL
"""
target_groups = pd.read_sql_query(sql_query,con)
target_groups.head()
```

```python
# first, do location

sql_query = """
SELECT labels.target_group, COUNT(DISTINCT events.user_id), events.location
FROM (
    SELECT user_id, 
        CASE WHEN jul > 0 AND aug IS NOT NULL THEN 'engaged' 
             WHEN jul > 0 AND aug IS NULL THEN 'stopped'
             ELSE NULL END AS target_group
    FROM (
        SELECT user_id,
                SUM(CASE WHEN month = 5.0 THEN monthly_events ELSE NULL END) AS may,
                SUM(CASE WHEN month = 6.0 THEN monthly_events ELSE NULL END) AS jun,
                SUM(CASE WHEN month = 7.0 THEN monthly_events ELSE NULL END) AS jul,
                SUM(CASE WHEN month = 8.0 THEN monthly_events ELSE NULL END) AS aug
         FROM(
            SELECT user_id, 
                    COUNT(event_name) AS monthly_events,
                    EXTRACT('month'  FROM occurred_at) AS month
             FROM events
             WHERE event_type = 'engagement'
             GROUP BY 1, 3
             ) sub
         GROUP BY 1
         ) sub_sub
) labels
LEFT JOIN events
 ON labels.user_id = events.user_id
 WHERE target_group IS NOT NULL
 GROUP BY labels.target_group, events.location
"""
loc = pd.read_sql_query(sql_query,con)
loc.head()
```

The emails table has an action column which contains info on how the user responds to the weekly email. Let's first make sure they're getting the email, and next see if they're even opening it, then finally see if they click through to the platform.

```python
# join the labels to the events table, and generate a weekly summary of email behavior by group.

sql_query = """
SELECT labels.target_group, 
    DATE_TRUNC('week', emails.occurred_at) AS week,
    COUNT(DISTINCT(emails.user_id)) AS sent_weekly_digest
FROM (
    SELECT user_id, 
        CASE WHEN jul > 0 AND aug IS NOT NULL THEN 'engaged' 
             WHEN jul > 0 AND aug IS NULL THEN 'stopped'
             ELSE NULL END AS target_group
    FROM (
        SELECT user_id,
                SUM(CASE WHEN month = 5.0 THEN monthly_events ELSE NULL END) AS may,
                SUM(CASE WHEN month = 6.0 THEN monthly_events ELSE NULL END) AS jun,
                SUM(CASE WHEN month = 7.0 THEN monthly_events ELSE NULL END) AS jul,
                SUM(CASE WHEN month = 8.0 THEN monthly_events ELSE NULL END) AS aug
         FROM(
            SELECT user_id, 
                    COUNT(event_name) AS monthly_events,
                    EXTRACT('month'  FROM occurred_at) AS month
             FROM events
             WHERE event_type = 'engagement'
             GROUP BY 1, 3
             ) sub
         GROUP BY 1
         ) sub_sub
) labels
LEFT JOIN emails
 ON labels.user_id = emails.user_id
 WHERE target_group IS NOT NULL 
 AND action = 'sent_weekly_digest'
 AND occurred_at > '2014-06-01'
 GROUP BY 1,2
"""
device = pd.read_sql_query(sql_query,con)
device.head()
```

```python
plt.figure(figsize=(10,5))
ax = sns.lineplot(x="week", y="sent_weekly_digest", hue = 'target_group', data=device)
plt.xticks(rotation=90)
plt.legend(bbox_to_anchor=(1, 1))
```

This chart is a little confusing, as it looks like the 'engaged' got more emails than the 'stopped'. That might be true, but that's because the engaged group, as I've defined it, acquired additional unique users through August. 

The take away here is that the 'stopped' group didn't see a drop in email deliveries. 
There isn't any indication to suggest this is a technical malfunction on the delivery side.

Let's see if both groups actually opened their emails.

```python
sql_query = """
SELECT labels.target_group, 
    DATE_TRUNC('week', emails.occurred_at) AS week,
    COUNT(DISTINCT(emails.user_id)) AS weekly_digest
FROM (
    SELECT user_id, 
        CASE WHEN jul > 0 AND aug IS NOT NULL THEN 'engaged' 
             WHEN jul > 0 AND aug IS NULL THEN 'stopped'
             ELSE NULL END AS target_group
    FROM (
        SELECT user_id,
                SUM(CASE WHEN month = 5.0 THEN monthly_events ELSE NULL END) AS may,
                SUM(CASE WHEN month = 6.0 THEN monthly_events ELSE NULL END) AS jun,
                SUM(CASE WHEN month = 7.0 THEN monthly_events ELSE NULL END) AS jul,
                SUM(CASE WHEN month = 8.0 THEN monthly_events ELSE NULL END) AS aug
         FROM(
            SELECT user_id, 
                    COUNT(event_name) AS monthly_events,
                    EXTRACT('month'  FROM occurred_at) AS month
             FROM events
             WHERE event_type = 'engagement'
             GROUP BY 1, 3
             ) sub
         GROUP BY 1
         ) sub_sub
) labels
LEFT JOIN emails
 ON labels.user_id = emails.user_id
 WHERE target_group IS NOT NULL 
 AND action = 'email_open'
 AND occurred_at > '2014-06-01'
 GROUP BY 1,2
"""
device = pd.read_sql_query(sql_query,con)
device.head()
```

```python
plt.figure(figsize=(10,5))
ax = sns.lineplot(x="week", y="weekly_digest", hue = 'target_group', data=device)
plt.xticks(rotation=90)
plt.legend(bbox_to_anchor=(1, 1))
```

```python
# by comapny

sql_query = """
SELECT labels.target_group, users.company_id, COUNT(DISTINCT(users.user_id))
FROM (
    SELECT user_id, 
        CASE WHEN jul > 0 AND aug IS NOT NULL THEN 'engaged' 
             WHEN jul > 0 AND aug IS NULL THEN 'stopped'
             ELSE NULL END AS target_group
    FROM (
        SELECT user_id,
                SUM(CASE WHEN month = 5.0 THEN monthly_events ELSE NULL END) AS may,
                SUM(CASE WHEN month = 6.0 THEN monthly_events ELSE NULL END) AS jun,
                SUM(CASE WHEN month = 7.0 THEN monthly_events ELSE NULL END) AS jul,
                SUM(CASE WHEN month = 8.0 THEN monthly_events ELSE NULL END) AS aug
         FROM(
            SELECT user_id, 
                    COUNT(event_name) AS monthly_events,
                    EXTRACT('month'  FROM occurred_at) AS month
             FROM events
             WHERE event_type = 'engagement'
             GROUP BY 1, 3
             ) sub
         GROUP BY 1
         ) sub_sub
) labels
LEFT JOIN users
 ON labels.user_id = users.user_id
 WHERE target_group IS NOT NULL
 GROUP BY 1,2
"""
company = pd.read_sql_query(sql_query,con)
company.head()
```

```python
# I tried plotting this and my computer almost crashed. Let's see how many companies we have
company.company_id.nunique()
```

```python
# OK, I'm going to take a different approach. I'm going to pivot the users/co by target_group.
# then I'm going to sum to get the total, and then calc the % stopped per company.
# I want to know if this is specific to a single company, or distributed.

company = company.pivot(index = 'company_id', columns='target_group', values='count')
company = company.reset_index(drop=True)
company['total'] = company.engaged + company.stopped
company['pct_stopped'] = 100*company.stopped / company.total
company.sort_values(by=['pct_stopped'], inplace=True, ascending=False)
company.head()
```

```python
# let's look at the stats on the per company % stopped.
company.pct_stopped.describe()
```

```python
company.shape
```

Summary stats indicate that within the companies that saw some loss of engagement, the lowest is 25%, and the highest is 82%. 
My take home from this is there's no single company that's causing this, say, by implementing a strict firewall or spam blocker. If there was, we'd be seeing 100% stoppage.
