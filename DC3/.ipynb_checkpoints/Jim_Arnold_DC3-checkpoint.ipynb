### Calculate Revenue

Labels looks to be about 66/33 split, which is what was expected.

Now, let's test if there's any difference in revenue.

To do this, I'm going to make a new column labeled 'revenue' - this will be the product of 'price' by 'converted'. 

Then, I'm going to perform a groupby and on 'test' and get the sum revenue by test group.
I also want the # of conversions, and the total group size. 
This will allow me to calc the conversion rate and the rev/person.

Quantity isn't provided, so I'm going to assume 1 purchase was made. This can be updated later if needed.

```python
# calc rev
test_results['revenue'] = test_results['price'] * test_results['converted']
```

#### define func to aggregate and compare metrics

```python
# make a new table with aggregate results
test_agg = test_results.groupby('test').agg(
    {'user_id': 'nunique', 'converted': 'sum', 'revenue': 'sum'})

# I'm going to call this a lot, so I'll make a function:


def per_user_stats(df):
    """takes an aggregated df and calculates the per visitor revenue and conversion rate, resets index"""
    df['RPV'] = df['revenue'] / df['user_id']
    df['conversion_rate'] = 100 * df['converted'] / df['user_id']
    df.reset_index(inplace=True)
    return df


test_agg = per_user_stats(test_agg)
```
