## Analysis/Modeling

### Steel

```python
# call model on Steel
steel = df[['ds', 'price_steel']].rename(columns={"price_steel":'y'})
m = Prophet()
m.fit(steel)
```

```python
# make future predictions 180 days into the future
future = m.make_future_dataframe(6, freq="MS")
forecast = m.predict(future)
fig = m.plot(forecast)
plt.xlabel('date')
plt.ylabel('price')
plt.title('price_steel')
```

```python
fig2 = m.plot_components(forecast)
plt.xticks(rotation=45)
```

#### Recommend

Based on component plots, prices are likely to increase after June. 

Recommend pursue 6 month contract in August to cover high fall and winter period. 

```python
# let's explore the historical changepoints to test this observation
from fbprophet.plot import add_changepoints_to_plot
fig = m.plot(forecast)
a = add_changepoints_to_plot(fig.gca(), m, forecast)
```

```python
# "borrowed" from https://pythondata.com/forecasting-time-series-data-with-prophet-part-2/

from datetime import date, timedelta


def plot_data(func_df, end_date):
    # find the 2nd to last row in the data. We don't take the last row because we want the charted lines to connect
    end_date = end_date - timedelta(days=180)
    # set up a mask to pull out the predicted rows of data.
    mask = (func_df.index > end_date)
    # using the mask, we create a new dataframe with just the predicted data.
    predict_df = func_df.loc[mask]

    # Now...plot everything
    fig, ax1 = plt.subplots()
    ax1.plot(df.y_orig)
    ax1.plot((np.exp(predict_df.yhat)), color='black', linestyle=':')
    ax1.fill_between(predict_df.index, np.exp(predict_df['yhat_upper']), np.exp(
        predict_df['yhat_lower']), alpha=0.5, color='darkgray')
    ax1.set_title('Sales (Orange) vs Sales Forecast (Black)')
    ax1.set_ylabel('Dollar Sales')
    ax1.set_xlabel('Date')

    # change the legend text
    L = ax1.legend()  # get the legend
    # change the legend text for 1st plot
    L.get_texts()[0].set_text('Actual Sales')
    # change the legend text for 2nd plot
    L.get_texts()[1].set_text('Forecasted Sales')


# call function to plot
plot_data(steel, steel.ds.max())
```

### Clean function

```python
# do that same, but for all remaining items
# write function to simplify


def material_prophet(df, material, n):
    """Takes as df of historical material price and forecasts prices n months into the future
    df is pandas dataframe with dates column labeled 'ds'
    material is string and column named in df
    n is integer for number of months to forecast into future

    returns 2 plots:
    1st shows historical data in blue with predicted trend in red
    2nd shows components of trend by year, weekday, and month"""

    # call model on material
    item = df[['ds', material]].rename(columns={material: 'y'})
    m = Prophet()
    m.fit(item)

    # make future predictions 180 days into the future
    future = m.make_future_dataframe(n, freq="MS")
    forecast = m.predict(future)

    # generate figure 1
    fig = m.plot(forecast)
    plt.xlabel('date')
    plt.ylabel('price')
    plt.title(material)

    # generate figure 2
    fig2 = m.plot_components(forecast)
    plt.xticks(rotation=45)
```

### Rebar

```python
# call on rebar
material_prophet(df, 'price_rebar', 6)
```

#### Recommend
Prices for rebar historically rise after June. Recommend pursuing contracts to cover winter months.

### Glass

```python
# call on glass
material_prophet(df, 'price_glass', 6)
```

#### Recommend
Glass prices are highest from March to May, puruse contracts in Jan.

### Concrete

```python
# call on concrete
material_prophet(df, 'price_concrete', 6)
```

#### Recommend
Concrete prices are highest from Feb to July, recommend puruse 6 month contracts in Jan.

### Plywood

```python
# call on plywood
material_prophet(df, 'price_plywood', 6)
```

#### Reccomend
